```python
# Loading data
train_data = pd.read_csv("/kaggle/input/cap-4611-2021-fall-assignment-02/train.csv")
test_data = pd.read_csv("/kaggle/input/cap-4611-2021-fall-assignment-02/eval.csv")
train_data.head()
```

```python
# Checking for missing values
train_data.isnull().sum()
```

```python
# Checking for missing values
test_data.isnull().sum()
```

```python
# Dropping irrelevant columns such as console, title, and id
X_train = train_data.drop(["console", "title", "id", "esrb_rating"], axis = 1)
X_test = test_data.drop(["console", "id"], axis = 1)
y_train = train_data["esrb_rating"]
```

```python
# Logistic regression model with evaluation
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
logreg = LogisticRegression()
logreg.fit(X_train,y_train)
y_pred = logreg.predict(X_test)
print(cross_val_score(logreg, X_train, y_train, cv=5).mean())
print(pd.DataFrame(y_pred).describe())
```

```python
# Support vector machine with evaluation and hyperparameters
from sklearn import svm
from sklearn.model_selection import GridSearchCV
param_grid = {"C": [0.1,1, 10, 100], "gamma": [1,0.1,0.01,0.001],"kernel": ["rbf", "poly", "sigmoid"]}
grid = GridSearchCV(svm.SVC(),param_grid,refit=True,verbose=1)
grid.fit(X_train,y_train)
estimator = grid.best_estimator_
grid_predictions = estimator.predict(X_test)
clf = svm.SVC()
clf.fit(X_train, y_train)
y_pred2 = clf.predict(X_test)
print(cross_val_score(clf, X_train, y_train, cv=5).mean())
print(cross_val_score(grid, X_train, y_train, cv=5).mean())
print(pd.DataFrame(y_pred2).describe())
```

```python
# Decision tree with evaluation
from sklearn.tree import DecisionTreeClassifier
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
y_pred3 = decision_tree.predict(X_test)
print(cross_val_score(decision_tree, X_train, y_train, cv=5).mean())
print(pd.DataFrame(y_pred3).describe())
```

```python
# Random forest with evaluation
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=10)
random_forest.fit(X_train, y_train)
y_pred4 = random_forest.predict(X_test)
print(cross_val_score(random_forest, X_train, y_train, cv=5).mean())
print(pd.DataFrame(y_pred4).describe())
```

```python
# KNeighbors with evaluation
from sklearn.neighbors import KNeighborsClassifier
neighbors = KNeighborsClassifier(n_neighbors=5)
neighbors.fit(X_train, y_train)
y_pred5 = neighbors.predict(X_test)
print(cross_val_score(neighbors, X_train, y_train, cv=5).mean())
print(pd.DataFrame(y_pred5).describe())
```
